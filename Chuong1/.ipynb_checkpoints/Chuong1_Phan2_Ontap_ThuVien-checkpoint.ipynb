## 2. On tap matplotlib
----
Thu vien matplotlib la thu vien chuyen su dung de truc quan hoa va ve cac bieu do

```python
import matplotlib.pyplot as plt
ten_khoa_x = ['K64','K65','K66','K67','K68']
so_hs_y = [450,570,1520, 820, 200]

# ve bieu do dang cot thang dung
plt.bar(ten_khoa_x, 
        so_hs_y)
plt.title('BIEU DO SO SINH VIEN THEO TUNG KHOA')
plt.xlabel('Khoa')
plt.ylabel('So luong')
plt.grid(ls='--')
plt.show()
```

```python
# ve bieu do dang cot nam ngang
plt.barh(ten_khoa_x, 
        so_hs_y,
       color='brown')
plt.title('BIEU DO SO SINH VIEN THEO TUNG KHOA')
plt.xlabel('So luong')
plt.ylabel('khoa')
plt.grid(ls='--')
plt.show()
```

```python
# ve bieu do dang duong
plt.plot(ten_khoa_x,
        so_hs_y,
        'r--s')
plt.grid('--')
plt.xlabel('Khoa')
plt.ylabel('So luong')
plt.title('BIEU DO SO SINH VIEN THEO TUNG KHOA')
plt.show()
```

```python
# ve bieu do tron
plt.pie(so_hs_y,
       labels=ten_khoa_x,
       autopct='%.2f%%',
       wedgeprops=dict(width=0.7,edgecolor='w'),)
plt.title('BIEU DO TY LE % SINH VIEN THEO TUNG KHOA')

plt.show()
```

```python
# Hien thi hai bieu do trong cung mot khung
plt.subplot(1,2,1)
# ve bieu do dang cot thang dung
plt.bar(ten_khoa_x, 
        so_hs_y)
plt.xlabel('Khoa')
plt.ylabel('So luong')
plt.grid(ls='--')

plt.subplot(1,2,2)
plt.pie(so_hs_y,
       labels=ten_khoa_x,
       autopct='%.2f%%',
       wedgeprops=dict(width=0.7,edgecolor='w'),)

plt.show()
```
